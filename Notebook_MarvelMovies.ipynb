```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
my_filepath="H:/Learning/KaggleMachineLearning/marvel_box_office_updated.csv"
my_data = pd.read_csv(my_filepath, index_col='Movie')
my_data['Average Score'] = ((my_data['IMDb Score'] * 10) + my_data['Meta Score'] + my_data['Tomatometer'] + my_data['Rotten Tomato Audience Score']) / 4
```

```python
# Create a plot
plt.figure(figsize=(16,20))
sns.barplot(y=my_data.index, x=my_data['Average Score'])
#sns.scatterplot(x=my_data['Release Month'], y=my_data['Tomatometer'])
```
